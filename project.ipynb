```python
df = pd.read_csv("annual-change-forest-area2.csv", low_memory=False)
dfcont = pd.read_csv("countryContinent.csv", encoding='ISO-8859-1')
dfEmissions = pd.read_csv("CO2-emission-by-countries.csv", encoding='ISO-8859-1')
```

```python
dfContinent = pd.merge(dfEmissions, dfcont, left_on="Country", right_on="country", how="left")
dfContinent.head()
```

```python
dfMerged = pd.merge(dfContinent, df, left_on=["Country", "Year"], right_on=["Entity", "Year"], how="left")
dfMerged.head()
```
